```python
# Step 2: Creation of a database
#conecto con la db, como no exite la DB a la que quiero conectar, la crea.
#creo cursor para podr usar comandos en la db


import sqlite3
connection = sqlite3.connect("../data/pelis_database.db")
cursor = connection.cursor()

#guardar df como tablas separadas 

pelis.to_sql("pelis_table", connection, if_exists = "replace", index = False)
creditos.to_sql("creditos_table", connection, if_exists = "replace", index = False)
```

```python
# Juntar las dos tablas para crear una tercera que contenga lo de las dos tablas combinadas
# usar title para el JOIN
# https://www.youtube.com/watch?v=4sul4F2PzZY&ab_channel=TMQuest


join_query = """
    SELECT *
    FROM pelis_table
    INNER JOIN creditos_table
    ON pelis_table.title = creditos_table.title;
"""

joint_data = pd.read_sql_query(join_query, connection)
connection.close()

joint_data = joint_data.loc[:, ~joint_data.columns.duplicated()]
joint_data.head()
```

```python
columns_to_keep = ['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']
joint_data = joint_data[columns_to_keep]
joint_data.head()
```

Step 3: Transform the data
As you can see, there are some JSON formatted columns. From each of the JSONs, select the name attribute and replace the genres and keywords columns. For the cast column, select the first three names.

The only columns left to modify are crew (team) and overview (summary). For the first column, convert it to contain the name of the director. For the second, convert it to a list.



```python
genres_data = joint_data.loc[0, 'genres']
print(genres_data)
```

```python
cast_data = joint_data.loc[0, 'cast']
print(cast_data)
```

```python
crew_data = joint_data.loc[0, 'crew']
print(crew_data)
```

```python
# Modificar data 
import json

def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value
    
joint_data["genres"] = joint_data["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
joint_data["keywords"] = joint_data["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)

joint_data["cast"] = joint_data["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)

joint_data["crew"] = joint_data["crew"].apply(lambda x: " ".join([crew_member['name'] for crew_member in load_json_safe(x) if crew_member['job'] == 'Director']))

joint_data["overview"] = joint_data["overview"].apply(lambda x: [x])

joint_data.head()
```

Once we have finished processing the columns and the recommendation model is not confused, for example, between Jennifer Aniston and Jennifer Conelly, we will remove the spaces between the words. Apply this function to the columns genres, cast, crew and keywords.

Finally, we will reduce our dataset by combining all of our previous converted columns into a single column called tags (which we will create). This column will now have all the elements separated by commas and then we will replace them with blanks. It should look something like this:

```python
joint_data["overview"] = joint_data["overview"].apply(lambda x: [str(x)])
joint_data["genres"] = joint_data["genres"].apply(lambda x: [str(genre) for genre in x])
joint_data["keywords"] = joint_data["keywords"].apply(lambda x: [str(keyword) for keyword in x])
joint_data["cast"] = joint_data["cast"].apply(lambda x: [str(actor) for actor in x])
joint_data["crew"] = joint_data["crew"].apply(lambda x: [str(crew_member) for crew_member in x])

joint_data["tags"] = joint_data["overview"] + joint_data["genres"] + joint_data["keywords"] + joint_data["cast"] + joint_data["crew"]
joint_data["tags"] = joint_data["tags"].apply(lambda x: ",".join(x).replace(",", " "))

joint_data.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)

joint_data.iloc[0].tags

```

To solve this problem we will create our own KNN. The first thing to do is to vectorize the text, following the same steps you learned in the previous lesson.

Once you have done that, we will have to choose a distance to compare texts. In this module we have seen a few, and the only one compatible with what we want to do is the cosine distance:

```python
# KNN mimplement modelo

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(joint_data["tags"])

model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
model.fit(tfidf_matrix)
```

Finally, we can design our similarity function based on the cosine distance. Our proposal is as follows:

```python
def get_movie_recommendations(movie_title):
    movie_index = joint_data[joint_data["title"] == movie_title].index[0]
    distances, indices = model.kneighbors(tfidf_matrix[movie_index])
    similar_movies = [(joint_data["title"][i], distances[0][j]) for j, i in enumerate(indices[0])]
    return similar_movies[1:]

input_movie = "Avatar"
recommendations = get_movie_recommendations(input_movie)
print("Film recommendations '{}'".format(input_movie))
for movie, distance in recommendations:
    print("- Film: {}".format(movie))
```
